# Using expenditures to estimate demand
$\text{predicted expenditures} := \hat{x} \\
\frac {\hat{x}}{q}  = \text{unit values} \\
p := \text{median unit values} \\
\hat{q} = \frac{\hat{x}}{p}$

Load the estimated result

```python
import cfe
result = cfe.regression.read_pickle('Bukidnon_estimates.pickle')
x_hat = result.predicted_expenditures().to_frame()
x_hat
```

Import helpful tools

```python
from eep153_tools.sheets import read_sheets
import numpy as np
import pandas as pd

def get_clean_sheet(key,sheet=None):

    df = read_sheets(key,sheet=sheet)
    df.columns = [c.strip() for c in df.columns.tolist()]

    df = df.loc[:,~df.columns.duplicated(keep='first')]   

    df = df.drop([col for col in df.columns if col.startswith('Unnamed')], axis=1)

    df = df.loc[~df.index.duplicated(), :]

    return df
```

Get observed quantities

```python
q = get_clean_sheet("https://docs.google.com/spreadsheets/d/1z35Kyga40hhoGBEQIY9jReY6z1wmnsnlUVw-pFEaBhg/edit#gid=1111889272", 
                   sheet="Quantities")
q.replace(0, np.nan, inplace=True)
q
```

Find unit values 

```python
unit_vals = x_hat / q
unit_vals
```
